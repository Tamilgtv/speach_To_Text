```python
times = {}
for i in df['results']:
    for wd in i['alternatives'][0]['timestamps']:
        times[(wd[1], wd[2])] = wd[0]

transcripts = {}
for j in df['speaker_labels']:
    speaker = j['speaker']
    if speaker in transcripts:
        transcripts[speaker].append(times[(j['from'], j['to'])])
    else:
        transcripts[speaker] = [times[(j['from'], j['to'])]]

print([{'speaker': k, 'transcript': ' '.join(transcripts[k])} for k in transcripts])
```

```python
frm = []
to = []
text = []
time = {}
for i in df['results']:
    for wd in i['alternatives'][0]['timestamps']:
        time[(wd[1],wd[2])] = wd[0]
        frm.append(wd[1])
        to.append(wd[2])
        text.append(wd[0])
```
